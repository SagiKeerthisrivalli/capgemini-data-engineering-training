```sql
CREATE TABLE employee_payments (
emp_id INT PRIMARY KEY,
emp_name VARCHAR(50),
department VARCHAR(30),
base_salary DECIMAL(10,2),
bonus DECIMAL(10,2),
joining_date DATE
);
INSERT INTO employee_payments VALUES
(1,'karthik','Data',75000.75,5000.50,'2019-03-15'),
(2,'veena','HR',65000.40,4000.25,'2021-06-20'),
(3,'ravi','Data',85000.90,6000.75,'2016-01-10'),
(4,'anil','Finance',70000.10,NULL,'2020-09-01'),
(5,'suresh','HR',60000.55,3000.30,'2022-11-25');
```

```sql
select upper(emp_name) as emp_upper,lower(emp_name) as emp_lower,INITCAP(emp_name)
from employee_payments
```

```sql
select INITCAP(emp_name),base_salary+coalesce(bonus,0) as total_income
from employee_payments
```

```sql
SELECT INITCAP(emp_name),ROUND(base_salary+coalesce(bonus,0)) as total_income
from employee_payments
```

```sql
select emp_name,extract(year from joining_date) as joining_date
from employee_payments
```

```sql
SELECT emp_name,round(datediff(current_date,joining_date)/365) as experience,
case
  when experience>7 then 'senior'
  when experience<7 and experience>4 then 'mid'
  else 'junior'
end as designation
FROM employee_payments;
```

```sql
CREATE TABLE orders_delivery (
order_id INT,
customer_name VARCHAR(50),
order_date DATE,
delivery_date DATE,
order_amount DECIMAL(10,2)
);
INSERT INTO orders_delivery VALUES
(101,'rajesh','2025-01-01','2025-01-05',12500.75),
(102,'meena','2025-01-10','2025-01-10',8400.40),
(103,'arun','2025-01-15','2025-01-20',15600.90),
(104,'pooja','2025-01-18',NULL,9200.10);
```

```sql
select upper(customer_name)
from orders_delivery
```

```sql
select ifnull(datediff(delivery_date,order_date),-1)
from orders_delivery
```

```sql
select ifnull(delivery_date,current_date)
from orders_delivery
```

```sql
select round(order_amount) as order_amount
from orders_delivery
```

```sql
select order_id,customer_name,order_date,delivery_date,
       case
           when delivery_date is null then 'Pending'
           when DATEDIFF(delivery_date, order_date) = 0 then 'Same-day'
           when DATEDIFF(delivery_date, order_date) > 3 then 'Delayed'
           else 'On-time'
       end as delivery_status
from orders_delivery;
```

```sql
CREATE TABLE customer_spending (
cust_id INT,
cust_name VARCHAR(50),
city VARCHAR(30),
purchase_amount DECIMAL(10,2),
purchase_date DATE
);
INSERT INTO customer_spending VALUES
(1,'amit','mumbai',12000.75,'2024-12-01'),
(2,'neha','delhi',8500.40,'2024-12-15'),
(3,'rohit','mumbai',15500.90,'2024-11-20'),
(4,'kavya','chennai',6000.10,'2024-10-05');
```

```sql
select INITCAP(cust_name) as cust_name ,city
from customer_spending
```

```sql
select cust_name,


date_format(purchase_date,'MMM') as month
from customer_spending
```

```sql
select cust_name, round(purchase_amount)
from customer_spending
```

```sql
SELECT cust_name,ABS(purchase_amount) AS purchase_amount
FROM customer_spending;
```

```sql
select cust_name,
  case 
   when purchase_amount>15000 then 'High spender'
   when purchase_amount<15000 and purchase_amount>8000 then 'Medium'
   else 'low'
  end type_of_customer
from customer_spending
```

```sql
CREATE TABLE subscriptions (
user_id INT,
user_email VARCHAR(100),
start_date DATE,
end_date DATE,
subscription_fee DECIMAL(10,2)
);
INSERT INTO subscriptions VALUES
(1,'karthik@gmail.com','2024-01-01','2025-01-01',12000.50),
(2,'veena@yahoo.com','2024-06-15','2024-12-15',8500.75),
(3,'ravi@hotmail.com','2023-03-01','2024-03-01',15000.90);
```

```sql
select user_id,substring(user_email,charindex('@',user_email)+1,len(user_email)) as domain
from subscriptions
```

```sql
select user_id,datediff(end_date,start_date) as duration_indays
from subscriptions
```

```sql
select
    user_id,
    user_email,
    format_number(subscription_fee, 2) AS formatted_fee
from subscriptions;
```

```sql
select 
    user_id,
    end_date,
    datediff(end_date, current_date()) AS remaining_days
from subscriptions;
```

```sql
select user_id,
case 
when datediff(current_date,end_date)>0 then 'expired'
when datediff(current_date,end_date)<30 then 'expiring soon'
else 'active'
end current_status
from subscriptions
```

```sql
CREATE TABLE loan_details (
loan_id INT,
customer_name VARCHAR(50),
loan_amount DECIMAL(12,2),
interest_rate DECIMAL(5,2),
loan_start DATE
);
INSERT INTO loan_details VALUES
(201,'suresh',500000.75,8.5,'2022-01-10'),
(202,'mahesh',750000.40,9.2,'2021-05-20'),
(203,'anita',300000.90,7.8,'2023-07-01');
```

```sql
select
    loan_id,
    customer_name,
    loan_amount*POWER(
        (1 +(interest_rate/100)/12),
        datediff(current_date,loan_start)/30
    ) AS amount
from loan_details;
```

```sql
select loan_id,customer_name,datediff(current_date,loan_start)/365 as yeasr_since_loan_start
from loan_details
```

```sql
select loan_id,customer_name,
case 
  when interest_rate>9 then 'high risk'
  when interest_rate<9 and interest_rate>4 then 'medium'
  else 'low'
  end risk_based_on_interest
from loan_details
```
