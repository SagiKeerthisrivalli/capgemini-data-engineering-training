```sql
select e.emp_id,e.emp_name,sum(s.amount)
from Employee e join sales as s on e.emp_id=s.emp_id
group by e.emp_id,e.emp_name
```

```sql
select s.product,sum(s.amount)
from Employee e join sales as s
on e.emp_id=s.emp_id
group by s.product
```

```sql
select s.product,avg(s.amount)
from Employee e join sales as s
on e.emp_id=s.emp_id
group by s.product
```

```sql
SELECT e.emp_name, COUNT(s.sales_id) AS total_sales
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.emp_name
HAVING COUNT(s.sales_id) > 2;
```

```sql
SELECT e.emp_name, sum(s.amount) AS total_sales,sum(e.salary) as total_salary
FROM Employee e
JOIN Sales s
ON e.emp_id = s.emp_id
GROUP BY e.emp_name
```

```sql
select e.emp_id,e.emp_name,count(distinct s.product) as uniqueproducts
from Employee e join sales s
  on e.emp_id = s.emp_id
group by e.emp_id, e.emp_name;
```

```sql
select e.emp_id,e.emp_name,max(s.amount) as maxsale
from Employee e join sales s
  on e.emp_id = s.emp_id
group by e.emp_id, e.emp_name;
```

```sql
select s.product,sum(s.amount)
from Employee e join sales s
  on e.emp_id = s.emp_id
group by s.product
having sum(s.amount)>50000
```

```sql
select e.department,avg(s.amount)
from Employee e join sales s
  on e.emp_id = s.emp_id
group by e.department
order by avg(s.amount)
limit 1
```

```sql
select e.department,max(s.amount)
from Employee e join sales s
  on e.emp_id = s.emp_id
group by e.department
order by max(s.amount)
limit 1
```

```sql
select e.emp_name,sum(s.amount)
from Employee e join sales s
  on e.emp_id = s.emp_id
group by e.emp_name
order by sum(s.amount)
limit 3
```

```sql
select year(e.joining_date),count(e.emp_id),avg(s.amount)
from Employee e join sales s
  on e.emp_id = s.emp_id
group by year(e.joining_date)
```
