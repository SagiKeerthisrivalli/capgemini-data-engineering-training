```sql
select Emp_id,emp_name,salary,
case 
when performance_rating = 'A' and Experience>=8 then 1.20*salary
when performance_rating = 'B' and Experience>=5 then 1.15*salary
when performance_rating = 'C ' then salary 
else 1.10*salary
end as new_salary
from Employee
```

```sql
select Emp_id,emp_name,salary,
case 
  when department='Finance' then
    case
        when performance_rating = 'A' then 0.20*salary
        when performance_rating = 'B' then 0.15*salary
        when performance_rating = 'C ' then 0.05*salary 
    end 
  when department='Engineering' then 
    case
        when performance_rating = 'A' then 0.18*salary
        when performance_rating = 'B' then 0.12*salary
        when performance_rating = 'C ' then 0.03*salary  
    end
  else 0.10*salary
end as new_salary
from Employee
```

```sql
select emp_name,salary,performance_rating,
case 
    when salary>80000 or performance_rating='A' then 'Highly performing'
    when salary<80000  and salary>50000 and performance_rating='B' then 'Mid performing'
    when salary<50000 or performance_rating='C' then 'low performing'
end Type_of_employee
from Employee
```

```sql
select emp_name,experience,department,
case
    when department='HR' then
      case 
        when experience<5 then 'high risk'
        when experience>5 then 'low risk'
        else 'medium risk'
      end
    when department='Engineering' or department='Finance' then
      case 
        when experience<8 then 'high risk'
        when experience>8 then 'low risk'
        else 'medium risk'
      end
    else 'medium risk'
end as Risk
from Employee
```

```sql
select emp_name,performance_rating,salary,
case 
  when performance_rating='A' then 
    case 
      when salary>80000 and experience>5 then 0.25*salary
      when salary<80000 and salary>50000 then 0.15*salary
    end
  when performance_rating='B' then
    case
      when experience>5 then 0.12*salary
      else 0.10*salary 
    end
  when performance_rating='C' then salary
end salary_hike
from employee     
```

```sql
select emp_name,experience,performance_rating,
case
    when department='HR' then
      case 
        when experience>5 and performance_rating='B' then 0.15*salary
        else 0.15*salary
      end
    when department='Finance' then
      case 
        when experience>10 and performance_rating='A' then 0.25*salary
        else 0.20*salary
      end
    else 0.08*salary
end as bonus
from Employee
```

```sql
select emp_name,experience,salary,performance_rating,
case
    when salary>70000 then
      case 
        when experience>8 and performance_rating='A' then 'top performer'
        when experience<=8 then 'mid perfomer'
        else 'no data given'
      end
    when salary<70000 and salary>50000 and performance_rating='A' then 'rising star'
    when salary<70000 and salary>50000 then 'average performer'
    when salary<50000 then 'low performer'
end as Risk
from Employee
```

```sql
select emp_name,experience,salary,performance_rating,
case
    when salary>90000 then
      case 
        when experience>10 then 0.65*salary
        else 0.70*salary
      end
    when salary<90000 and salary>60000  then
      case 
        when experience>5 then 0.75*salary
        else 0.80*salary 
        end
    when salary<60000 then 0.85*salary 
end as tax_money
from Employee
```

```sql
select emp_name,experience,salary,performance_rating,
case
    when performance_rating='A' then
      case 
        when experience>7 and salary>75000 then 'Eligible for Senior Role'
        else 'Eligible for Junior Role'
      end
    when performance_rating='B' then 
      case 
        when experience>5 then 'eligible for consideration'
        else 'not eligible'
      end
    when performance_rating='C' then 'not eligible'
end as Promotion_possibility
from Employee
```
