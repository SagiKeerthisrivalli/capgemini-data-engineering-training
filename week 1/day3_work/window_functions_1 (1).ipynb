###ROW_NUMBER QUESTIONS

```sql
SELECT emp_id, emp_name, salary,
ROW_NUMBER() OVER (ORDER BY salary DESC) AS row_num
FROM employees;
```

```sql
SELECT emp_id, emp_name, salary,
ROW_NUMBER() OVER (partition by department ORDER BY salary DESC) AS row_num
FROM employees;
```

```sql

select emp_id,emp_name,join_date,
ROW_NUMBER()  over(order by join_date desc) as row_num
from employees;
```

```sql

select emp_id,emp_name,join_date,
ROW_NUMBER()  over(order by join_date asc) as row_num
from employees;
```

```sql
select emp_id,emp_name,join_date,
ROW_NUMBER()  over(order by salary asc) as row_num
from employees;
```

```sql
select emp_id,emp_name,join_date,
ROW_NUMBER()  over(order by emp_name asc) as row_num
from employees;
```

###RANK() ONLY Questions

```sql
select emp_name,salary,
RANK() over(order by salary desc) as rank
from employees
```

```sql
select emp_name,salary,department,
RANK() over(partition by department order by salary desc) as rank
from employees
```

```sql
select emp_name,join_date,
RANK() over(order by join_date desc) as rank
from employees
```

```sql
select emp_name,department,join_date,
RANK() over(partition by department order by join_date desc) as rank
from employees
```

```sql
select emp_name,
RANK() over(order by emp_name asc) as rank
from employees
```

###DENSE_RANK() ONLY Questions

```sql
select emp_name,
DENSE_RANK() over(order by emp_name asc) as rank
from employees
```

```sql
select emp_name,salary,
DENSE_RANK() over(order by salary asc) as rank
from employees
```

```sql
select emp_name,department,salary,
DENSE_RANK() over(partition by department order by salary asc) as rank
from employees
```

```sql
select emp_name,department,join_date,
DENSE_RANK() over(partition by department order by join_date asc) as rank
from employees
```
