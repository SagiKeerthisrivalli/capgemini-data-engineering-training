##CTE

```python
customers.createOrReplaceTempView("customers")
policies.createOrReplaceTempView("policies")
claims.createOrReplaceTempView("claims")
```

 compute risk score

```python
%sql
with prem_cte as(
    select customer_id,sum(premium) as total_p
    from policies 
    group by customer_id
),
claim_cte as(
  select p.customer_id,sum(c.claim_amount) as total_c
  from policies p join claims c on p.policy_id=c.policy_id
  group by p.customer_id
),
risk_cte as(
  select p.customer_id,coalesce(c.total_c,0) as total_c,
  coalesce(p.total_p,0) as total_p
  from prem_cte p left join claim_cte c
  on p.customer_id=c.customer_id
)
SELECT * FROM risk_cte;
```

Find top 3 risky customers per city

```python
%sql
with prem_cte as(
    select customer_id,sum(premium) as total_p
    from policies 
    group by customer_id
),
claim_cte as(
  select p.customer_id,sum(c.claim_amount) as total_c
  from policies p join claims c on p.policy_id=c.policy_id
  group by p.customer_id
),
risk_cte as(
  select p.customer_id,coalesce(c.total_c,0) as total_c,
  coalesce(p.total_p,0) as total_p,(COALESCE(c.total_c,0) / p.total_p) AS risk_score
  from prem_cte p left join claim_cte c
  on p.customer_id=c.customer_id
),
city_cte as(
  select r.customer_id,cu.city,r.risk_score,rank() over(partition By cu.city order by r.risk_score desc) as rank
from risk_cte r join customers cu
on r.customer_id=cu.customer_id
)
SELECT *
FROM city_cte
WHERE rank <= 3;

```

customers whose claims are increasing month-over-month

```python
%sql
with monthly_claims AS (
    select p.customer_id,
           date_format(c.claim_date, 'yyyy-MM') AS month,
           SUM(c.claim_amount) AS total_claim
    FROM claims c
    JOIN policies p
    ON c.policy_id = p.policy_id
    GROUP BY p.customer_id, month
),

lag_cte AS (
    SELECT customer_id,
           month,
           total_claim,
           LAG(total_claim) OVER (
               PARTITION BY customer_id
               ORDER BY month
           ) AS prev_claim
    FROM monthly_claims
)

SELECT *
FROM lag_cte
WHERE total_claim > prev_claim;
```
